### Wine: alcohol content prediction based on multiple parameters

Data set from https://archive.ics.uci.edu/ml/datasets/wine

```python
# imports 
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix
import sklearn.linear_model
import sklearn.preprocessing
```

```python
# data import 
wine = pd.read_csv("winequality-all.csv", comment="#")

# quick look at data 
wine.head()
wine.info()
wine.dtypes

```

```python
# chaning color object into category 
wine.color = wine.color.astype("category")
```

```python
# checking number of red / white wine within the set 
wine.color.value_counts()

# separatation of dataset into red and white subsets 
white = wine[wine.color == 'white']
red = wine[wine.color == 'red']

# dropping color column, since no longer needed
white = white.drop('color', axis=1)
red = red.drop('color', axis=1)
```

```python
# dividing white subset into dependent (desribed) and independent variables  
X = white.loc[:,white.columns !='alcohol']   # independent variables, i.e. everything but alcohol content 
y = white.loc[:,'alcohol']                   # dependent variable, i.e. alcohol content 
  
# dividing X and y into train and test sets 
X_train, X_test, y_train, y_test = sklearn.model_selection.train_test_split(X,y, train_size = 0.8)
```

```python
# looking into correlatation between variables 
corr_p = white.corr('pearson') # pearson correlation coeff.
#tworzymy macierz trójkątną i wyświetlamy wspóczynnik korelacji większy od 0.5
# corr_p = corr_p.where(np.tril(np.ones(corr_p.shape),k=-1).astype(np.bool)) # polowka macierzy korelacji
corr_p_half = corr_p.where(np.tril(np.ones(corr_p.shape),k=-1).astype(bool)) # half is enough, matrix is symmetric 
corr_p_half
```

```python
# list of correlations above abs(0.5)
corr_p_tri = corr_p.where(np.triu(np.ones(corr_p.shape, dtype=bool), k=1)).stack().sort_values()
corr_p_tri[abs(corr_p_tri)>0.5]
```

```python
# correlation matrix visualised 
with sns.axes_style('white'):
    f,ax = plt.subplots(figsize=(15,10))
    ax= sns.heatmap(corr_p_half,vmin=-1, vmax=1, cmap='RdBu', linewidth =.1, annot = True)
    plt.rc('xtick', labelsize=9) 
    plt.rc('ytick', labelsize=9)
    ax.tick_params(axis = 'x', labelrotation = 45)
    ax.xaxis.tick_top()    
```

#### Machine learning models - linear regression

```python
# function training and testing particular model
def fit_regression(X_train, X_test, y_train, y_test):
    r = sklearn.linear_model.LinearRegression()
    r.fit(X_train, y_train)
    y_train_pred = r.predict(X_train)
    y_train_pred = r.predict(X_test)
    r2 = sklearn.metrics.r2_score
    mse = sklearn.metrics.mean_squared_error
    mae = sklearn.metrics.mean_absolute_error
    return {
        "r_score_u": r2(y_train, y_train_pred),
        "r_score_t": r2(y_test, y_test_pred),
        "MSE_u": mse(y_train, y_train_pred),
        "MSE_t": mse(y_test, y_test_pred),
        "MAE_u": mae(y_train, y_train_pred),
        "MAE_t": mae(y_test, y_test_pred)
    }
```
